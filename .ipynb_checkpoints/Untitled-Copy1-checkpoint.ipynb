## Using kfold cross validation to ensure reliability and stability.

```python
# defining the functions to be used in the cross validation
def train_model(df_train,y_train,C):
    train_dict= df_train[categorical+numerical].to_dict(orient='records')
    dv=DictVectorizer()
    x_train=dv.fit_transform(train_dict)
    model=LogisticRegression(penalty='l2',solver='liblinear',C=C,class_weight='balanced',max_iter=1000,random_state=42)
    model.fit(x_train,y_train)
    return dv,model
```

```python
dv,model=train_model(df_train,y_train,C=1)
```

```python
def predict(df_val,model,dv):
    val_dict= df_val[categorical+numerical].to_dict(orient='records')
    x_val=dv.transform(val_dict)
    y_scores=model.predict_proba(x_val)[:,1]
    return y_scores
```

```python
y_scores=predict(df_val,model,dv)
```

```python
from sklearn.model_selection import KFold
```

```python
kfold=KFold(n_splits=5,shuffle=True,random_state=42)
```

```python
for train_idx,val_idx in kfold.split(df_fulltrain):
    df_train=df_fulltrain.iloc[train_idx]
    df_val=df_fulltrain.iloc[val_idx]
    y_train= df_train.isfraud.values
    y_val=df_val.isfraud.values

    del df_train['isfraud']
    del df_val['isfraud']
    dv,model=train_model(df_train,y_train,C=1)
    y_scores=predict(df_val,model,dv)
    auc_log=roc_auc_score(y_val,y_scores)
    pr_auc_log=average_precision_score(y_val,y_scores)
    print(f'auc_log:{auc_log:.2f},pr_auc_log:{pr_auc_log:.2f}')
    
```

###### Crossvalidation confirmed that while ROCAUC remained consistently high (0.931.00), PRAUC varied more widely (0.450.86). This highlights the challenge of imbalanced datasets: ROCAUC can overstate performance, while PRAUC provides a more realistic measure of fraud detection reliability. The lowest PR_AUC happened because the model founf it difficult to detect a fraud case in that split due to the extremely high class imbalance.

**I chose not to aggregate out-of-fold predictions for Logistic Regression because its fold-level performance was already stable. Across all folds, ROC-AUC remained consistently high (0.931.00), and PR-AUC showed moderate variation but no major outliers. This consistency suggests that threshold behavior is predictable, so aggregation would add complexity without significantly improving reliability**

## Hyperparameter tuning

```python
c_values=[0.000001,0.00001,0.0001,0.001,0.01,1]
results={}
for C in c_values:
    scores=[]
    for train_idx,val_idx in kfold.split(df_fulltrain):
        df_train=df_fulltrain.iloc[train_idx]
        df_val=df_fulltrain.iloc[val_idx]
        y_train= df_train.isfraud.values
        y_val=df_val.isfraud.values
    
        del df_train['isfraud']
        del df_val['isfraud']
        dv,model=train_model(df_train,y_train,C=C)
        y_scores=predict(df_val,model,dv)
        auc_log=roc_auc_score(y_val,y_scores)
        pr_auc_log=average_precision_score(y_val,y_scores)
        score=[auc_log,pr_auc_log]
        scores.append(score)
        results[C]=np.round(np.mean(scores),2)
y_pred=y_scores>=0.985       
print('results:',results)
precision_log= precision_score(y_val,y_pred)
recall_log=recall_score(y_val,y_pred)
print()
print(f'precision score for logistic regression model is {precision_log:.3}. Recall score for this same model is {recall_log:.3}')
```

From the scores above the best hyperparameter is 1 anyway as used initially so no need to change anything

## Training a random forest model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
def train_model_rf(df_train,y_train,n=10,d=3,m=1):
    train_dict= df_train[categorical+numerical].to_dict(orient='records')
    dv=DictVectorizer()
    x_train=dv.fit_transform(train_dict)
    model_rf=RandomForestClassifier(n_estimators=n,random_state=42,class_weight='balanced',max_depth=d,min_samples_leaf=m)
    model_rf.fit(x_train,y_train)
    return dv,model_rf

def predict(df_val,model_rf,dv):
    val_dict= df_val[categorical+numerical].to_dict(orient='records')
    x_val=dv.transform(val_dict)
    y_scores_rf=model_rf.predict_proba(x_val)[:,1]
    return y_scores_rf

dv,model_rf=train_model_rf(df_train,y_train,n=10,d=3,m=1)
y_scores_rf=predict(df_val,model_rf,dv)
y_pred_rf=y_scores_rf>=0.6
```

```python

for train_idx,val_idx in kfold.split(df_fulltrain):
    df_train=df_fulltrain.iloc[train_idx]
    df_val=df_fulltrain.iloc[val_idx]
    y_train= df_train.isfraud.values
    y_val=df_val.isfraud.values

    del df_train['isfraud']
    del df_val['isfraud']
    dv,model_rf=train_model_rf(df_train,y_train,n=10,d=3,m=1)
    y_scores_rf=predict(df_val,model_rf,dv)
    
    auc_rf=roc_auc_score(y_val,y_scores_rf)
    pr_auc_rf=average_precision_score(y_val,y_scores_rf)
    print(f'auc:{auc_rf:.2f},pr_auc_rf:{pr_auc_rf:.2f}')
```

###### Tuning the parameter for the number of trees. 

```python

%%time
kfold=KFold(n_splits=2,shuffle=True,random_state=42)
# applying parameters to the model
results=[]
for n in range (10,150,20):
     print(f"Training with n_estimators={n}") # I printed this to track the progress of my model
     for train_idx,val_idx in kfold.split(df_fulltrain):
        df_train=df_fulltrain.iloc[train_idx]
        df_val=df_fulltrain.iloc[val_idx]
        y_train= df_train.isfraud.values
        y_val=df_val.isfraud.values
    
        del df_train['isfraud']
        del df_val['isfraud']
        dv,model_rf=train_model_rf(df_train,y_train,n,d=3,m=1)
        y_scores_rf=predict(df_val,model_rf,dv)
        auc_rf=roc_auc_score(y_val,y_scores_rf)
        pr_auc_rf=average_precision_score(y_val,y_scores_rf)
        scores=[auc_rf,pr_auc_rf]
        result=[n,np.mean(scores)]
        results.append(result)
    
df_results=pd.DataFrame(results,columns=['no.est','avg']) 
df_results
```

```python
plt.plot(df_results['no.est'],df_results['avg'])
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Average CV Score (ROC-AUC or PR-AUC)')
plt.title('Random Forest Performance vs Number of Trees')
plt.grid(True)
plt.show()
```

## Training an XGBoost classifier

```python
from xgboost import XGBClassifier
```

```python
def train_model_xg(df_train,y_train,n=10,d=3,m=1,g=0,s=1,c=0.5,e=0.1):
    train_dict= df_train[categorical+numerical].to_dict(orient='records')
    dv=DictVectorizer()
    x_train=dv.fit_transform(train_dict)
    model_xg=XGBClassifier(n_estimators=n,random_state=42,max_depth=d,min_child_weight=m,gamma=g,scale_pos_weight=s,colsample_bytree=c,eta=e)
    model_xg.fit(x_train,y_train)
    return dv,model_xg

def predict(df_val,model_xg,dv):
    val_dict= df_val[categorical+numerical].to_dict(orient='records')
    x_val=dv.transform(val_dict)
    y_scores_xg=model_xg.predict_proba(x_val)[:,1]
    return y_scores_xg

dv,model_xg=train_model_xg(df_train,y_train,n=10,d=3,m=1,g=0,s=1,c=0.5,e=0.1)
y_scores_xg=predict(df_val,model_xg,dv)
y_pred_xg=y_scores_xg>=0.6
```

```python
%%time
# Tuning hyperparameters
results=[]

for d in (5,11,20):
    for m in (1,5):
        for g in (0,1,5):
            for s in (10,25,40):
                for c in (0.7,1):
                    for e in (0.05,0.1):
                        for train_idx,val_idx in kfold.split(df_fulltrain):
                            df_train=df_fulltrain.iloc[train_idx]
                            df_val=df_fulltrain.iloc[val_idx]
                            y_train= df_train.isfraud.values
                            y_val=df_val.isfraud.values
                        
                            del df_train['isfraud']
                            del df_val['isfraud']
                            dv,model_xg=train_model_xg(df_train,y_train,n=90,d=d,m=m,g=g,s=s,c=c,e=e)
                            y_scores_xg=predict(df_val,model_xg,dv)
                            auc_xg=roc_auc_score(y_val,y_scores_xg)
                            pr_auc_xg=average_precision_score(y_val,y_scores_xg)
                            parameters=[d,m,g,s,c,e]
                            scores=[auc_xg,pr_auc_xg]
                            result=parameters+[np.mean(scores)]
                            results.append(result)
                         
df_results=pd.DataFrame(results,columns=['max_depth','min_child_weight','gamma','scale_pos_weight','col_sample_bytree','eta','avg_of_auc_prauc']) 
df_results
                             
                        

        
```

```python

y_val_all=[]
y_scores_all=[]
scores=[]
for train_idx,val_idx in kfold.split(df_fulltrain):
    df_train=df_fulltrain.iloc[train_idx]
    df_val=df_fulltrain.iloc[val_idx]
    y_train= df_train.isfraud.values
    y_val=df_val.isfraud.values
    
    del df_train['isfraud']
    del df_val['isfraud']
    dv,model_xg=train_model_xg(df_train,y_train,n=90,d=5,m=1,g=0,s=10,c=0.7,e=0.1)
    y_scores_xg=predict(df_val,model_xg,dv)
    y_val_all.append(y_val)
    y_scores_all.append(y_scores_xg)
    auc_xg=roc_auc_score(y_val,y_scores_xg)
    pr_auc_xg=average_precision_score(y_val,y_scores_xg)
    score=[auc_xg,pr_auc_xg]
    scores.append(score)

y_val_all=np.concatenate(y_val_all)
y_scores_all=np.concatenate(y_scores_all)
y_pred_xg=y_scores_xg>=0.5
print(confusion_matrix(y_val,y_pred_xg))
precision_xg= precision_score(y_val,y_pred_xg)
print('XGBoost precision score',precision_xg)
recall_xg= recall_score(y_val,y_pred_xg)
print('XGBoost recall score:',recall_xg)
```

#### Tuning threshold for xgboost classifier.
I am tuning the threshold based on the best recall where precision is also good. In other words I am prioritising recall based on my bussiness goal.

```python
# Tuning the threshold using the precision recall curve
precision_xg,recall_xg,threshold=precision_recall_curve(y_val_all,y_scores_all)
for p,r,t in zip(precision_xg[:-1],recall_xg[:-1],threshold):
    if p >= 0.4:
        print(f'precision={p:.2f},recall={r:.2f},threshold={t:.3f}')
```

```python
# plotting the precision recall curve for the xgboost model
plt.figure(figsize=(6, 6))
plt.plot(recall_xg, precision_xg, marker='.', label='XGBoost')
plt.xlabel('Recall_xg')
plt.ylabel('Precision_xg')
plt.title('PrecisionRecall Curve for XGBoost')
plt.legend()
plt.grid(True)
plt.show()
```

###### After tuning the threshold I ended up picking a threshold where recall (prioritised metric) is very high when precision is still good. A threshold of 0.024 gave a recall score of 0.85 and a precision score of 0.5

```python
# Final xg boost model
y_val_all=[]
y_scores_all=[]
for train_idx,val_idx in kfold.split(df_fulltrain):
    df_train=df_fulltrain.iloc[train_idx]
    df_val=df_fulltrain.iloc[val_idx]
    y_train= df_train.isfraud.values
    y_val=df_val.isfraud.values
    
    del df_train['isfraud']
    del df_val['isfraud']
    dv,model_xg=train_model_xg(df_train,y_train,n=90,d=5,m=1,g=0,s=10,c=0.7,e=0.1)
    y_scores_xg=predict(df_val,model_xg,dv)
    y_val_all.append(y_val)
    y_scores_all.append(y_scores_xg)
    auc_xg=roc_auc_score(y_val,y_scores_xg)
    pr_auc_xg=average_precision_score(y_val,y_scores_xg)
    score=[auc_xg,pr_auc_xg]
    scores.append(score)

y_val_all=np.concatenate(y_val_all)
y_scores_all=np.concatenate(y_scores_all)
y_pred_all=y_scores_all>=0.038
print('General confusion matrix',confusion_matrix(y_val_all,y_pred_all))
precision_all_xg= precision_score(y_val_all,y_pred_all)
print('XGBoost General precision score',precision_all_xg)
recall_all_xg= recall_score(y_val_all,y_pred_all)
print('XGBoost General recall score:',recall_all_xg)
print('XGboost General average precision (pr_auc) score:', average_precision_score(y_val_all,y_scores_all))
print('XGBoost Genera roc_auc score',roc_auc_score(y_val_all,y_scores_all))
```
